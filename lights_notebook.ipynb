# Lights Notebook

```python
# %matplotlib widget

# Import 
import lights
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

big_circle = (25.4/2)*1e-2  # Radius of big circle [m]
small_circle = 5*1e-2       # Radius of small circle [m]
lights_dia = 1e-2           # Lights diameter [m]
n_circles = 4               # Number of circles
n_lights = 70               # Number of lights

[circles_radius, n_lights_circle] = lights.pattern(big_circle, small_circle, n_circles, n_lights)  # Calculate radius of the cicles and number of lights per cicle
_ = lights.circles_plot(circles_radius, n_lights_circle, lights_dia)
```

## Table of measurements

```python
cols = ['circle radius [cm]', 'number of lights', 'angle [deg]']
thetas = 360/ n_lights_circle
int_lights = n_lights_circle.astype(int)
data = np.array((circles_radius*1e2, int_lights, thetas))
measurements = pd.DataFrame(data.T, columns=cols)

pd.set_option('display.precision',2)
measurements.head()
```

## Save figure

Save the printed figure to use as a template with real size dimensions

```python
m = 1/2.54e-2     # Convert [in] to [m]
sheet_size = np.array([circles_radius[-1], circles_radius[-1]])*2*1.1

fig = lights.circles_plot(circles_radius*m, n_lights_circle, lights_dia*m, fig_size=sheet_size*m)
fig.savefig('lights_tempate.pdf')
```
